```python
import numpy as np
from ase.build import molecule
from ase.geometry.analysis import Analysis

from ase.io import Trajectory

structures = Trajectory('moldyn_300K.traj')
structures = [s for s in structures]
ana = Analysis(structures)
RDF = ana.get_rdf(rmax=3.8864017382470784, nbins=100)



### Library for plotting results
import matplotlib.pyplot as plt

plt.plot(RDF[-1])
plt.show()
```

```python
#For the last 
filtered=np.average(RDF[150:200], axis = 0)
print(filtered)
```
